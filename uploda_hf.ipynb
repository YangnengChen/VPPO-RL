```python
from huggingface_hub import HfApi, upload_folder
import os

# --- 1. 定义你的路径 ---

# 你本地模型文件夹的路径
local_folder_path = "/data3/cyn/projects/VPPO-RL/checkpoints/3b_vppo/perc0.4_advsc0.9_pen0.06_ep2_rollout5_MINI_ROLLOUT_BATCH_SIZE384/global_step_202/actor/huggingface"

# --- 2. 定义 Hugging Face Hub 上的目标仓库 ---

# 你想上传到的仓库 ID (格式: "ynchen/3b_vppo_perc0.4_advsc0.9_pen0.06_ep2_rollout5_MINI_ROLLOUT_BATCH_SIZE384")
# ！！请务必修改为你自己的用户名和仓库名！！
repo_id = "ynchen11/3b_vppo_perc0.4_advsc0.9_pen0.06_ep2_rollout5_MINI_ROLLOUT_BATCH_SIZE384" 

# 例如: "cyn/vppo-3b-global-step-202-actor"

# --- 3. (可选) 检查路径是否存在 ---
if not os.path.exists(local_folder_path):
    print(f"错误：路径不存在: {local_folder_path}")
else:
    print(f"准备从 {local_folder_path} 上传到 {repo_id}...")

    try:
        # --- 4. 执行上传 ---
        # upload_folder 会上传指定文件夹中的 *所有* 文件
        
        # HfApi().upload_folder(...) 提供了更多控制
        api = HfApi()
        api.upload_folder(
            folder_path=local_folder_path,
            repo_id=repo_id,
            repo_type="model",         # 仓库类型 (可以是 'model', 'dataset', 'space')
            create_repo=True,          # 如果 repo_id 不存在，自动创建它
            commit_message="Upload PyTorch actor model from global_step_202" # 提交信息
        )

        print(f"\n上传成功！")
        print(f"查看你的模型: https://huggingface.co/{repo_id}")

    except Exception as e:
        print(f"上传失败: {e}")
```
